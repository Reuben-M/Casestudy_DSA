## Standard Scaling:

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data1=scaler.fit_transform(data)
```

## K-Means Clustering

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state= 42)
    kmeans.fit(data1)
    wcss.append(kmeans.inertia_)
    print(kmeans.inertia_)
plt.plot(range(1,11),wcss)  
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel('WCSS')
plt.show()
    
```

From the graph by applying Elbow method we can see that the optimum value of K is 3

```python
#K-Means
kmeans=KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans=kmeans.fit_predict(data1)
type(data1)
```

```python
y_kmeans
```

```python
#converting into a dataframe
y_kmeans=pd.DataFrame(y_kmeans)
y_kmeans=y_kmeans.rename(columns={0:'cluster'})
```

```python
#joining with the original data
data2=pd.concat([data,y_kmeans],axis=1)
```

```python
data2.head()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(18,8))
sns.scatterplot(data=data2, x='Alcohol', y='Malic_Acid', hue='cluster', ax=ax[0])
sns.scatterplot(data=data2, x='Ash', y='Ash_Alcanity', hue='cluster', ax=ax[1])
sns.scatterplot(data=data2, x='Total_Phenols', y='Color_Intensity', hue='cluster', ax=ax[2])
plt.show()
```

Cluster 0 has high Malic Acid Content as well as color Intensity but low amount of Phenols(Phenols are chemicals that affect  the taste, color, and mouthfeel (i.e., texture) of wine.


Cluster 1 has high Phenol content as well as Alcohol and color intensity. This indicates that presence of more Phenols 
affects Alcohol level of a wine.

All the clusters only have a moderate amount of Ash and alkanity in them which is actually good. (Ash is simply the inorganic matter left after evaporation and incineration, the alkalinity of ash determines how basic (as opposed to acidic) the ash in a wine is).

# Agglomerative Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(data1, method='ward'))
plt.title('Dendogram')
plt.show()
```

From the dendogram its again clear that the optimum value of clusters is 3

```python
from sklearn.cluster import AgglomerativeClustering
ahc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage ='ward')
y_ahc = ahc.fit_predict(data1)
y_ahc
```

```python
y_ahc = pd.DataFrame(y_ahc)
y_ahc = y_ahc.rename(columns={0:'ahc_cluster'})
```

```python
data3=pd.concat([data,y_ahc],axis=1)
data3.head()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(18,8))
sns.scatterplot(data=data3, x='Alcohol', y='Malic_Acid', hue='ahc_cluster', ax=ax[0])
sns.scatterplot(data=data3, x='Ash', y='Ash_Alcanity', hue='ahc_cluster', ax=ax[1])
sns.scatterplot(data=data3, x='Total_Phenols', y='Color_Intensity', hue='ahc_cluster', ax=ax[2])
plt.show()
```

Cluster 0 has high Malic Acid Content as well as color Intensity but low amount of Phenols(Phenols are chemicals that affect  the taste, color, and mouthfeel (i.e., texture) of wine.


Cluster 1 has high Phenol content as well as Alcohol and color intensity. This indicates that presence of more Phenols 
affects Alcohol level of a wine.

All the clusters only have a moderate amount of Ash and alkanity in them which is actually good. (Ash is simply the inorganic matter left after evaporation and incineration, the alkalinity of ash determines how basic (as opposed to acidic) the ash in a wine is).
