## ● Check whether all the zones generate the same amount of sales.

Here we can emply ONE_WAY ANOVA to check whether qll the zones generate the same amount of sales.

We can formulate the Hypothesis as-

Null hypothesis H(0) :- All the zones generate same amount of sales

Alternative hypothesis H(a) :- All the zones generate different amount of sales

Significance level, alpha value= 0.05

if p_value < alpha_value We reject the null hypothesis

if P_value > alpha_value We accept the null hypothesis 


```python
f_stat,p_value=stat.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'], axis=0)
```

```python
f_stat
```

```python
p_value
```

Here we see that the P value that we got is lower than 0.05 hence we reject the NUll Hypothesis.
That means all the zones generate different amounts of sale!
